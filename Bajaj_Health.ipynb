Share percentage of missing values, for the columns firstName, lastName and DOB from patientDetails.

```python
import pandas as pd
def extract_names_dob(data):

  extracted_data = []
  for record in data:
    patient_details = record.get("patientDetails", {})
    extracted_data.append({
        "firstName": patient_details.get("firstName", ""),
        "lastName": patient_details.get("lastName", ""),
        "DOB": patient_details.get("birthDate", "").split('T')[0] if patient_details.get("birthDate") else ""
    })

  df = pd.DataFrame(extracted_data)
  return df

df_names_dob = extract_names_dob(data)
print(df_names_dob)
```

```python
df_names_dob.head()
```

```python
# prompt: find the missing value percentage of each coloumn individually upto two decimal places from df_names_dob

# Calculate missing value percentages
missing_percentages = {}
for col in ["firstName", "lastName", "DOB"]:
  missing_count = df_names_dob[col].apply(lambda x: x == "" or x.isspace()).sum()
  missing_percentages[col] = round((missing_count / len(df_names_dob)) * 100, 2)

print("\nMissing Value Percentages:")
print(missing_percentages)
```

```python
import pandas as pd
def extract_gender(data):
  extracted_data = []
  for record in data:
    patient_details = record.get("patientDetails", {})
    extracted_data.append({
        "gender": patient_details.get("gender", "")
    })
  df = pd.DataFrame(extracted_data)
  return df

df_gender = extract_gender(data)
df_gender.head()
```

```python
mode_gender = df_gender['gender'].mode()[0]

df_gender['gender'].fillna(mode_gender, inplace=True)

female_count = df_gender[df_gender['gender'] == 'F'].shape[0]
total_count = df_gender.shape[0]
female_percentage = (female_count / total_count) * 100

print("\nPercentage of Female Gender (After Imputation): {:.2f}%".format(female_percentage))
```

```python

from datetime import datetime

def calculate_age(dob_str):
  if dob_str:
    dob = datetime.strptime(dob_str, '%Y-%m-%d')
    today = datetime.today()
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    return age
  else:
    return None

df_names_dob['age'] = df_names_dob['DOB'].apply(calculate_age)

def categorize_age(age):
  if age is None:
    return None
  elif age <= 12:
    return "Child"
  elif age <= 19:
    return "Teen"
  elif age <= 59:
    return "Adult"
  else:
    return "Senior"

df_names_dob['ageGroup'] = df_names_dob['age'].apply(categorize_age)

df = df_names_dob
df
```

```python
adult_count = df[df['ageGroup'] == 'Adult']['ageGroup'].count()
print("Count of Adults:", adult_count)
```

```python
# prompt: extract medicineName and store in df

import pandas as pd
def extract_medicine_names(data):
  extracted_data = []
  for record in data:
    medications = record.get("medications", [])
    for medication in medications:
      extracted_data.append({
          "medicineName": medication.get("medicineName", "")
      })
  df = pd.DataFrame(extracted_data)
  return df

df_medicines = extract_medicine_names(data)
print(df_medicines)
```
